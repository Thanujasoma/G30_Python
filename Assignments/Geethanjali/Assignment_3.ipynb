```python
#1
import random

def choose_word():
    words = ['python', 'hangman', 'computer', 'programming', 'game']
    return random.choice(words)

def display_word(word, guessed_letters):
    displayed_word = ''
    for letter in word:
        if letter in guessed_letters:
            displayed_word += letter
        else:
            displayed_word += '_'
    return displayed_word

def hangman():
    word = choose_word()
    guessed_letters = []
    max_attempts = 6

    print("Welcome to hangman!")
    print("Try to guess the word.")

    while True:
        print("\n" + display_word(word, guessed_letters))

        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You already guessed letter.")
        else:
            guessed_letters.append(guess)
            if guess in word:
                print("Correct.")
            else:
                print("Incorrect.")
                max_attempts -= 1

        if max_attempts == 0:
            print("You lost. The word was:", word)
            break
        elif all(letter in guessed_letters for letter in word):
            print("Congratulations! You guessed the word:", word)
            break

hangman()
```

```python
#2
def grocery_budget_tracker(budget):
    total_spent = 0
    while True:
        item = input("Enter item name (or 'done' to finish): ")
        if item.lower() == 'done':
            break
        price = float(input("Enter item price: $"))
        total_spent += price

    print("\nTotal spent: ${:.2f}".format(total_spent))
    if total_spent > budget:
        overspent = total_spent - budget
        print("You've overspent by ${:.2f}!".format(overspent))
    else:
        saved = budget - total_spent
        print("You've saved ${:.2f}!".format(saved))

budget = float(input("Enter your grocery budget: $"))
grocery_budget_tracker(budget)
```

```python
#3
def check_password_strength(password):
    length = len(password)
    if length < 8:
        return "Weak! Password length should be at least 8 characters."

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(not char.isalnum() for char in password)

    if not (has_upper and has_lower and has_digit and has_special):
        return "Weak! Password should contain at least one uppercase letter, one lowercase letter, one digit, and one special character."

    return "Strong! Your password meets all complexity criteria."

def suggest_password_improvements(password):
    suggestions = []

    length = len(password)
    if length < 8:
        suggestions.append("Consider using a longer password (at least 8 characters).")

    if not any(char.isupper() for char in password):
        suggestions.append("Include at least one uppercase letter.")

    if not any(char.islower() for char in password):
        suggestions.append("Include at least one lowercase letter.")

    if not any(char.isdigit() for char in password):
        suggestions.append("Include at least one digit.")

    if not any(not char.isalnum() for char in password):
        suggestions.append("Include at least one special character.")

    return suggestions

password = input("Enter your password: ")

strength_result = check_password_strength(password)
print(strength_result)

if "Weak" in strength_result:
    improvements = suggest_password_improvements(password)
    print("\nSuggestions for improvement:")
    for suggestion in improvements:
        print("- " + suggestion)
```

```python
#5
def quiz():
    questions = [
        "What is the capital of France?",
        "Who wrote 'Romeo and Juliet'?",
        "What is the powerhouse of the cell?",
        "What year did the first moon landing occur?",
        "How many days are in a leap year?"
    ]

    answers = [
        "Paris",
        "William Shakespeare",
        "Mitochondria",
        "1969",
        "366"
    ]

    feedback = [
        "Correct! Paris is the capital of France.",
        "Correct! William Shakespeare wrote 'Romeo and Juliet'.",
        "Correct! Mitochondria is the powerhouse of the cell.",
        "Correct! The first moon landing occurred in 1969.",
        "Correct! A leap year has 366 days."
    ]

    score = 0
        for i in range(len(questions)):
        print(questions[i])
        user_answer = input("Your answer: ")
        if user_answer.lower() == answers[i].lower():
            print(feedback[i])
            score += 1
        else:
            print("Incorrect. answer is:", answers[i])

    print("Quiz completed! You scored", score, "out of", len(questions))
    if score == len(questions):
        print("Congratulations! You got a perfect score!")
    elif score >= len(questions) // 2:
        print("Well done! You did pretty well.")
    else:
        print("Keep practicing! You can do better.")

quiz()
```

```python
#6
def currency_converter():
    conversion_rates = {
        'USD': 1.01,
        'EUR': 0.88,
        'GBP': 0.90,
        'JPY': 109.10,
    }
    amount = float(input("Enter your amount: "))
    from_currency = input("Enter the currency you have (e.g., USD, EUR, GBP, JPY): ").upper()
    to_currency = input("Enter the currency you want to convert to (e.g., USD, EUR, GBP, JPY): ").upper()

    if from_currency in conversion_rates and to_currency in conversion_rates:
        converted_amount = amount * conversion_rates[to_currency] / conversion_rates[from_currency]
        print(f"{amount} {from_currency} is equal to {converted_amount} {to_currency}")
    else:
        print(" Please enter valid currency codes.")

currency_converter()
```

```python
#4
income = 0
expenses = 0
total_expenses = 0
categories = {}

income = float(input("Enter your income: $"))

while True:
    category = input("Enter expense category (or type 'done' to finish): ")
    if category.lower() == 'done':
        break
    amount = float(input(f"Enter amount spent on {category}: $"))
    total_expenses += amount
    categories[category] = categories.get(category, 0) + amount

remaining_balance = income - total_expenses

print("\n--- Summary ---")
print(f"Total Income: ${income}")
print(f"Total Expenses: ${total_expenses}")
print(f"Remaining Balance: ${remaining_balance}")

if categories:
    print("\n--- Expense Breakdown ---")
    for category, amount in categories.items():
        print(f"{category}: ${amount}")
```
